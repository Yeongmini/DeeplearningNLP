8. 워드 임베딩(Word Enbbeding)

텍스트를 컴퓨터가 이해하고 효율적으로 처리하게 하기 위해서는 컴퓨터가 이해할 수 있도록 텍스트를 적절히 숫자로 변환해야 한다.

워드 임베딩이란 단어를 벡터로 표현하는 방법으로, 단어를 밀집 표현으로 변환합니다. 희소 표현, 밀집 표현, 그리고 워드 임베딩에 대한 개념을 학습합니다.

1. 희소 표현(Sparse Representation)

원-핫 인코딩을 통해 나온 원-핫 벡터들은 표현하고자 하는 단어의 인덱스만 1이고 나머지는 다 0으로 처리한다. \
이렇게 벡터 또는 행렬의 값 대부분이 0으로 표현되는 방법을 희소 표현이라고 한다.(원-핫 벡터는 희소 벡터이다.)

희소 벡터의 문제점은 단어의 개수가 늘어나면 벡터의 차원이 한없이 커진다는 것이다.\
ex) 코퍼스에 단어가 10,000개 있다면 벡터의 차원은 10,000이어야만 한다.\
이러한 벡터 표현은 공간적 낭비를 불러일으킨다. 이러한 공간적 낭비는 원-핫 벡터뿐만 아니라 DTM과 같은경우에도 나타난다.

2. 밀집 표현(Dense Representation)

희소 표현의 반대되는 표현으로 밀집 표현이 있다. 밀집 표현은 벡터의 차원을 단어 집합의 크기로 상정하지 않는다. 사용자가 설정한 값으로 모든 단어의 벡터 표현의 차원을 맞춘다. \
희소 표현 ex) 강아지 = [0 0 0 1 0 0 ...중략... 0] # 이때 1 뒤의 0의 수는 9995개. 차원은 10,000 \
밀집 표현 ex) 강아지 = [0.2 1.8 1.1 -2.1 1.1 2.8 ....] # 벡터의 차원은 128(사용자 설정)

3. 워드 임베딩(Word Embedding)

단어를 밀집 벡터의 형태로 표현하는 방법을 워드 임베딩이라고 한다. 그리고 이 밀집 벡터를 워드 임베딩 과정을 통해 나온 결과라고 하여 임베딩 벡터라고도 한다.\
워드 임베딩 방법론으로는 LSA, Word2Vec, FastText, Glove 등이 있다.
